```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
# change Flight number to integer data type fill the missing two values by increasing by 10 from previous value & 
df['FlightNumber']  = pd.Series(df['FlightNumber'] , dtype = pd.Int64Dtype())
df.isna()

for i in range(0 , len(df['FlightNumber'])-1):
    if(pd.isnull(df['FlightNumber'][i])):
        df['FlightNumber'][i] = df['FlightNumber'][i-1]+10
        
df
        
```

```python
#split into to and from columns and 
tempdf = df
tempdf[['From','To']] = tempdf['From_To'].str.split('_',expand=True)


tempdf
```

```python
#change to title case for to and from columns
tempdf['From'] = tempdf['From'].str.title()
tempdf['To'] = tempdf['To'].str.title()
tempdf
```

```python
#delete from_To column from df
df = df.drop('From_To', 1)
df['From'] = tempdf['From']
df['To'] = tempdf['To']
df
```

```python
#find maxlength of list in the column
maxlen = 0
for i in df['RecentDelays']:
    if(len(i) > maxlen):
        maxlen = len(i)

#create empty columns
#print(maxlen)
for i in range(0, maxlen):
    df['delay_' + str(i)] = np.NaN

#loop through column and add elements by each row with max len value
for i in range(len(df['RecentDelays'])):
    for j in range(0, len(df['RecentDelays'][i])):
        df['delay_' + str(j)][i] = int(df['RecentDelays'][i][j])
df
```
